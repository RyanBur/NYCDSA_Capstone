```python
import numpy as np
import pandas as pd
```

```python
zip_codes = pd.read_csv('acs2018.csv')
```

```python
zip_codes = zip_codes[['geo_id']]
zip_codes
```

```python
hospitals = pd.read_csv('Hospitals_HIFL.csv')
```

```python
hospitals = hospitals.groupby('ZIP').agg('count')['X'].reset_index()
hospitals.columns=['zip_code', 'HIFL_hospital_count']
hospitals
```

```python
transport = pd.read_csv('Public_Transit_Stations_HIFL.csv')
```

```python
transport = transport.groupby('ZIP').agg('count')['X'].reset_index()
transport.columns=['zip_code', 'HIFL_transit_station_count']
transport
```

```python
hosp_transit_df = zip_codes.merge(hospitals, how='left', left_on='geo_id', right_on='zip_code').drop('zip_code', axis=1).fillna(0)
hosp_transit_df['HIFL_hospital_count'].value_counts()
```

```python
hosp_transit_df = hosp_transit_df.merge(transport, how='left', left_on='geo_id', 
                                        right_on='zip_code').drop('zip_code', axis=1).fillna(0)
hosp_transit_df['HIFL_transit_station_count'].value_counts()
```
