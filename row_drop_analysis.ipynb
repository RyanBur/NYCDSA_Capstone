```python
# Fill interior NaN's
zillow_trim = pd.merge(zillow_trim.iloc[:,0:7], zillow_trim.iloc[:,8:].interpolate(limit_area='inside', axis=1), left_index=True, right_index=True)
```

```python
zillow_trim = zillow_trim.reset_index(drop=True)
```

```python
zillow_trim.iloc[-50:-20, 6:]
```

```python
# Fill in the beginning NaN's. 
for row in range(len(zillow_trim)):
    i = 7 # first rent column
    while True:
        if pd.isnull(zillow_trim.iloc[row, i]):
            i+=1
        else:
            break
    i -= 1
    while i > 6:
        zillow_trim.iloc[row, i] = (zillow_trim.iloc[row, i+12] * .9615)
        i-=1
```

```python
zillow_trim.iloc[-50:-20, 6:]
```

```python
zillow_trim.iloc[-20:, -10:]
```

```python
# Fill in the ending NaN's. 
for row in range(len(zillow_trim)):
    i = 62 # last rent column
    while True:
        if pd.isnull(zillow_trim.iloc[row, i]):
            i-=1
        else:
            break
    i += 1
    while i < 63:
        zillow_trim.iloc[row, i] = (zillow_trim.iloc[row, i-12] * 1.04)
        i+=1
```
