```python
import numpy as np
import pandas as pd
```

# Gov Building

```python
gov_data = pd.read_csv('Major_State_Government_Buildings.csv')
```

```python
gov_data.columns
```

```python
gov_data
```

```python
gov_data = gov_data.groupby('ZIP').count().reset_index()
```

```python
gov_data
```

```python
gov_data = gov_data.drop([
    'X',
    'Y',
    'TELEPHONE',
    'ADDRESS2',
    'COUNTY',
    'DIRECTIONS',
    'GEOPREC',
    'PHONELOC',
    'LONGITUDE',
    'LATITUDE',
    'ZIP4'
], axis=1)
```

```python
gov_data.describe()
```

```python
gov_data = gov_data.drop([
    'FID',
    'NAME',
    'ADDRESS',
    'CITY',
    'STATE',
    'FIPS',
    'NAICSCODE',
    'NAICSDESCR',
    'AGENCIES',
    'AGENCY_PH',
    'NUM_AGENCY',
], axis=1)
```

```python
gov_data['HIFL_govbldg'] = gov_data['UNIQUE_ID']
gov_data = gov_data.drop('UNIQUE_ID', axis=1)
```

```python
HIFL_govbldg = gov_data
```

```python
HIFL_govbldg
```

# College/University

```python
uni_data = pd.read_csv('Colleges_and_Universities_Campuses.csv')
```

```python
uni_data.columns
```

```python
uni_data = uni_data.groupby('ZIP').count().reset_index()
```

```python
uni_data
```

```python
uni_data.describe()
```

```python
uni_data = uni_data[['ZIP', 'UNIQUEID']]
```

```python
uni_data['HIFL_uni'] = uni_data['UNIQUEID'] 
uni_data = uni_data.drop('UNIQUEID', axis=1)
```

```python
HIFL_uni = uni_data
```

```python
HIFL_uni
```

# merge df

```python
data = pd.read_csv('acs2018.csv')
```

```python
zip_data = data[['geo_id']]
```

```python
zip_data
```

```python
uni_gov_df = zip_data.merge(HIFL_uni, how='left', left_on='geo_id', right_on='ZIP').drop('ZIP', axis=1).fillna(0)
```

```python
uni_gov_df['HIFL_uni'].value_counts()
```

```python
uni_gov_df = uni_gov_df.merge(HIFL_govbldg, how='left',
                              left_on='geo_id', right_on='ZIP').drop('ZIP', axis=1).fillna(0)
```

```python
uni_gov_df['HIFL_govbldg'].value_counts()
```

```python
uni_gov_df
```

```python
uni_gov_df.to_csv('HIFL_uni_gov_features.csv', index=False)
```
